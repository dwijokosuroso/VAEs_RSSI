```python
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

# using some dummy data for this example
xs = np.arange(0,10,1)
ys = np.random.normal(loc=3, scale=0.4, size=10)

plt.barh(xs,ys)

# zip joins x and y coordinates in pairs
for x,y in zip(xs,ys):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (y,x), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,0), # distance from text to points (x,y)
                 ha='left',
                va='center') # horizontal alignment can be left, right or center

plt.show()
```

```python
df = pd.read_csv('BarGraphVAE2.csv')
```

```python
df_new = df.set_index("Location")
```

```python
df_ap1 = df_new.iloc[:5,:3]
```

```python
df_ap1
```

```python
# ECTICON-22 GRAPH

font = {'size': 16}
  
# using rc function
plt.rc('font', **font)

legend_properties = {'weight':'bold'}
fontweight = 'bold'
fontproperties = {'weight' : fontweight}
yticks = df.iloc[:5,0]


ax = df_ap1.plot.barh(width = 0.75, figsize=(17, 18))
rects = ax.patches
ax.set_xticklabels(ax.get_xticks(), fontproperties)
ax.set_yticklabels(yticks, fontproperties)

for rect in rects:
    x_value = rect.get_width()
    y_value = rect.get_y() + rect.get_height() / 2
    space = 5
    
    label = "{:.2f}".format(x_value)
    plt.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(space, 0),          # Horizontally shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        va='center', 
        fontweight='bold',
        ha='left') 
    
    
plt.title('Epoch vs. Average RSSI Error Values (-dBm)', fontweight='bold')
plt.ylabel('Sampled Location', fontweight='bold')
plt.legend(prop=legend_properties)
plt.savefig("Barh_AP1_revision.png", bbox_inches='tight')
plt.savefig('average error comaprison.png')
plt.show()
```
